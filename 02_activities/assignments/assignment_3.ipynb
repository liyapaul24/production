# Load the data

Place the files in the ../../05_src/data/fires/ directory and load the appropriate file. 

```python
# Load the libraries as required.
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PowerTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
# Load data
columns = [
    'coord_x', 'coord_y', 'month', 'day', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain', 'area' 
]
fires_dt = (pd.read_csv('../../05_src/data/fires/forestfires.csv', header = 0, names = columns))
fires_dt.info()
```

# Get X and Y

Create the features data frame and target data.

```python
# Define the features
X = fires_dt.drop('area', axis=1)  

print(X.head())
```

```python
# Define the target
Y = fires_dt['area']

print(Y.head())
```

# Preprocessing

Create two [Column Transformers](https://scikit-learn.org/stable/modules/generated/sklearn.compose.ColumnTransformer.html), called preproc1 and preproc2, with the following guidelines:

- Numerical variables

    * (Preproc 1 and 2) Scaling: use a scaling method of your choice (Standard, Robust, Min-Max). 
    * Preproc 2 only: 
        
        + Choose a transformation for any of your input variables (or several of them). Evaluate if this transformation is convenient.
        + The choice of scaler is up to you.

- Categorical variables: 
    
    * (Preproc 1 and 2) Apply [one-hot encoding](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html) where appropriate.


+ The only difference between preproc1 and preproc2 is the non-linear transformation of the numerical variables.
    


### Preproc 1

Create preproc1 below.

+ Numeric: scaled variables, no other transforms.
+ Categorical: one-hot encoding.

```python
# Define the columns
numeric_features = ['coord_x', 'coord_y', 'ffmc', 'dmc', 'dc', 'isi', 'temp', 'rh', 'wind', 'rain']
categorical_features = ['month', 'day']

# Preprocessing for numerical and categorical features
numeric_transformer_1 = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='infrequent_if_exist')


# Column Transformer for preproc1
preproc1 = ColumnTransformer(
    transformers=[
        ('numeric', numeric_transformer_1, numeric_features),
        ('categorical', categorical_transformer, categorical_features)
    ]
)

preproc1
```

### Preproc 2

Create preproc1 below.

+ Numeric: scaled variables, non-linear transformation to one or more variables.
+ Categorical: one-hot encoding.

```python
# Non-linear transformation 
numeric_transformer_2 = Pipeline(steps=[
    ('scaler', StandardScaler()),
    ('power', PowerTransformer())  
])

# Column Transformer for preproc2
preproc2 = ColumnTransformer(
    transformers=[
        ('numeric', numeric_transformer_2, numeric_features),
        ('categorical', categorical_transformer, categorical_features)
    ]
)

preproc2
```

## Model Pipeline


Create a [model pipeline](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html): 

+ Add a step labelled `preprocessing` and assign the Column Transformer from the previous section.
+ Add a step labelled `regressor` and assign a regression model to it. 

## Regressor

+ Use a regression model to perform a prediction. 

    - Choose a baseline regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Choose a more advance regressor, tune it (if necessary) using grid search, and evaluate it using cross-validation.
    - Both model choices are up to you, feel free to experiment.

```python
# Pipeline A = preproc1 + baseline
# Define the Baseline and Advanced Regressors
baseline_regressor = Ridge()
advanced_regressor = RandomForestRegressor()

pipe_a = Pipeline(steps=[
    ('preprocessing', preproc1),
    ('regressor', baseline_regressor)
])

pipe_a
```

```python
# Pipeline B = preproc2 + baseline
pipe_b = Pipeline(steps=[
    ('preprocessing', preproc2),
    ('regressor', baseline_regressor)
])

pipe_b
```

```python
# Pipeline C = preproc1 + advanced model
pipe_c = Pipeline(steps=[
    ('preprocessing', preproc1),
    ('regressor', advanced_regressor)
])

pipe_c
```

```python
# Pipeline D = preproc2 + advanced model
pipe_d = Pipeline(steps=[
    ('preprocessing', preproc2),
    ('regressor', advanced_regressor)
])
    
pipe_d   
    
```

# Tune Hyperparams

+ Perform GridSearch on each of the four pipelines. 
+ Tune at least one hyperparameter per pipeline.
+ Experiment with at least four value combinations per pipeline.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Hyperparameter grid for Ridge (baseline model)
ridge_params = {'regressor__alpha': [0.01, 0.1, 1, 10]}
```

```python
# Hyperparameter grid for RandomForestRegressor (advanced model)

rf_params = {
    'regressor__n_estimators': [50, 100, 150, 200], 
    'regressor__max_depth': [None, 10, 20, 30]      
}
```

```python
# GridSearch on each pipeline 
grid_a = GridSearchCV(pipe_a, param_grid=ridge_params, cv=5, scoring='neg_mean_squared_error')
grid_a.fit(X, Y)
grid_b = GridSearchCV(pipe_b, param_grid=ridge_params, cv=5, scoring='neg_mean_squared_error')
grid_b.fit(X, Y)
grid_c = GridSearchCV(pipe_c, param_grid=rf_params, cv=5, scoring='neg_mean_squared_error')
grid_c.fit(X, Y)
grid_d = GridSearchCV(pipe_d, param_grid=rf_params, cv=5, scoring='neg_mean_squared_error')
grid_d.fit(X, Y)
```

```python
# Collect best scores and parameters
print("Pipeline A - Best Score:", -grid_a.best_score_, "Best Params:", grid_a.best_params_)
print("Pipeline B - Best Score:", -grid_b.best_score_, "Best Params:", grid_b.best_params_)
print("Pipeline C - Best Score:", -grid_c.best_score_, "Best Params:", grid_c.best_params_)
print("Pipeline D - Best Score:", -grid_d.best_score_, "Best Params:", grid_d.best_params_)
```

```python
scores = {
    'Pipeline A': -grid_a.best_score_,
    'Pipeline B': -grid_b.best_score_,
    'Pipeline C': -grid_c.best_score_,
    'Pipeline D': -grid_d.best_score_,
}

best_pipeline = min(scores, key=scores.get)
best_model = scores[best_pipeline]

print(f"The best performing model is {best_pipeline} with: {best_model}")
```

The best performing model is Pipeline A with score of 4278.655 and best parameters {'regressor__alpha': 10}.

# Explain

+ Use SHAP values to explain the following only for the best-performing model:

    - Select an observation in your test set and explain which are the most important features that explain that observation's specific prediction.

    - In general, across the complete training set, which features are the most and least important.

+ If you were to remove features from the model, which ones would you remove? Why? How would you test that these features are actually enhancing model performance?

```python
import shap

# Fit the pipeline A on the training data
best_performing_model.fit(X_train, Y_train)

# Transform the test data using the preprocessing step from your pipeline
data_transform = best_performing_model.named_steps['preprocessing'].transform(X_test)

# Initialize the SHAP explainer for a linear model
explainer = shap.explainers.Linear(
    best_performing_model.named_steps['regressor'], 
    data_transform,
    feature_names=best_performing_model.named_steps['preprocessing'].get_feature_names_out()
)

# Calculate SHAP values for the transformed data
shap_values = explainer(data_transform)
```

```python
shap.plots.waterfall(shap_values[1])
```

```python
shap.plots.beeswarm(shap_values)
```

1. the waterfall plot shows that the most important positive contributors to the model's prediction are DMC (8.94), coord_x (6.08), and temp (2.08). This indicates that higher values of DMC (moisture content), the x-coordinate of the fire location, and temperature are associated with larger predicted burned areas. On the other hand, the most significant negative contributors are DC (-2.60), rain (-2.40), and RH (-2.36), suggesting that higher values of DC (fuel moisture), rainfall, and relative humidity reduce the predicted area burned, likely because these factors inhibit fire growth. Additionally, 20 other features collectively contribute -2.84, implying a cumulative negative impact, and it may be worth investigating which specific features within this group are influencing the prediction.

2. Based on the beeswarm plot, across the complete training set, the most important features are DMC, coord_x, and DC. DMC has the highest SHAP values, indicating it plays a significant role in predicting the burned area, with large variations in its values affecting the prediction. Similarly, coord_x (the fire's x-coordinate) and DC (fuel moisture content) show strong impacts, with their SHAP values varying across the dataset. On the other hand, the least important features are Day_thu and Day_fri, which have relatively low SHAP values, suggesting that the day of the week does not significantly influence the model's predictions about the size of the fire.

3. If I were to remove features, I would consider eliminating Day_thu and Day_fri, as they show low SHAP values, indicating minimal impact on the model’s predictions. To test if removing these features affects performance, I would first exclude them from both the training and test datasets, then retrain the model on the reduced data. I would evaluate the model's performance using metrics like RMSE or R-squared and compare the results with the original model. If the reduced model performs similarly or better, it suggests these features were not adding significant value, and removing them simplifies the model. If performance declines, it indicates these features had some predictive power.
